```python
import pandas as pd
from nltk import word_tokenize
from nltk import sent_tokenize
import re
from deep_translator import GoogleTranslator
```

```python
# Read in the info to the DataFrame
# Each row represents one video with the VIDEO_TAGS column being 
# a list of all tags for that video.
df = pd.read_pickle('Data_Out/auto_channels_data.pkl')
df
```

```python
# This takes about 40 hours to run
reg1 = re.compile('[\_]') # match underscore
reg2 = re.compile('[\s]') # match whitespace

for index, row in df.iterrows():
    tempList = []
    for tag in row.VIDEO_TAGS:
        tag = re.sub(reg1, ' ', tag).lower() # Replace underscores with spaces  
        try:
            trans = GoogleTranslator(source='auto', target='en').translate(tag)
            if tag != trans: # If the translation actually changed the tag
                print(index, ' ',tag, '---->', trans)
            else: pass
            tag_ = re.sub(reg2, '_', trans) # Replace spaces in tags with underscore
            tempList.append(tag_)
        except:
            tag_ = re.sub(reg2, '_', tag) # Replace spaces in tags with underscores
            tempList.append(tag_)
    row.VIDEO_TAGS = tempList
```

```python
# Save the new DataFrame
df.to_pickle('Data_Out/auto_channels_translated.pkl')
```
